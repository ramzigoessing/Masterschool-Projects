```python
from transformers import pipeline

# Load the sentiment analysis pipeline
classifier = pipeline("sentiment-analysis")

# Input text
text = "I had an amazing experience at the restaurant!"

# Get the classification result
result = classifier(text)
print(result)
```

```python
from transformers import pipeline

# Load the summarization pipeline
summarizer = pipeline("summarization")

# Input text
text = """
The Hugging Face Transformers library provides access to pre-trained models for tasks like text classification,
summarization, question answering, and more. It simplifies the integration of advanced NLP into your projects.
"""

# Get the summary
summary = summarizer(text, max_length=50, min_length=25, do_sample=False)
print(summary)
```

```python
from transformers import pipeline

# Load the question-answering pipeline
qa_pipeline = pipeline("question-answering")

# Input context and question
context = """
The Eiffel Tower is a wrought-iron lattice tower in Paris, France. It is named after the engineer Gustave Eiffel,
whose company designed and built the tower.
"""
question = "Who designed the Eiffel Tower?"

# Get the answer
answer = qa_pipeline(question=question, context=context)
print(answer)
```

```python
from transformers import pipeline

# Load the translation pipeline
translator = pipeline("translation_en_to_fr")

# Input text
text = "Hugging Face is an amazing library for NLP tasks!"

# Get the translated text
translation = translator(text, max_length=50)
print(translation)
```
